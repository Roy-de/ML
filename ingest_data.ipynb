```python
import sqlalchemy as sq
import psycopg2
```

```python
try:
   engine = sq.create_engine("postgresql://postgres:7934@localhost/Real estate",echo=True)
   engine.connect()
   print("Connected Successfully")
   
except IOError:
    print("ERROR: ",IOError)   
```

```python
#Create tables according to the type of data you want to input to your database
metadata = sq.MetaData()

# database name
real_estate = sq.Table(
    'Real estate',
    metadata,
    sq.Column('Number',sq.BIGINT,primary_key=True),
    sq.Column('Transaction_date',sq.FLOAT),
    sq.Column('House_age',sq.FLOAT),
    sq.Column('distance_to_MRT',sq.FLOAT),
    sq.Column('No_of_stores',sq.INTEGER),
    sq.Column('Latitude',sq.FLOAT),
    sq.Column('Longitude',sq.FLOAT),
    sq.Column('House_Price',sq.FLOAT),
)
#Create Real estate tables
metadata.create_all(engine)
```

```python
import pandas as pd
import numpy as np
from time import time
```

```python
dataframe = pd.read_csv(r'/home/roy/Downloads/datasets-main/Real estate.csv',iterator=True,chunksize=1000)
```

```python
while True:
    try:    
        start_t = time()
        df_net = next(dataframe)
        dbname = "Real estate"
        df_net.head(n=0).to_sql(name=dbname,con= engine,if_exists='replace')
        df_net.to_sql(name=dbname,con=engine,if_exists='append')
        end = time()
        print("inserted another chunk, took %.3f seconds" % (end - start_t))

    except StopIteration:
      print("Finished ingesting data")
      break
```
